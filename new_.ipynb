# 링크 아이디 편집하기

```python
link_node=pd.read_excel('tm_link.xlsx')
```

```python
from pyproj import Transformer
transformer = Transformer.from_crs("epsg:5181", "epsg:4166")
```

```python
#x=link_node['x'][0]
#y=link_node['y'][0]

transformer.transform(197521.131469,452291.143788)
```

```python
x_r=[]
y_r=[]
for i in range(47019):
    x_,y_=link_node['x'][i],link_node['y'][i]
    x_,y_=transformer.transform(y_, x_)
    x_r.append(x_)
    y_r.append(y_)
```

```python
link_node['위도']=x_r
```

```python
link_node['경도']=y_r
```

```python
link_node
```

```python
link_node=link_node.groupby(link_node['링크아이디']).mean()
```
